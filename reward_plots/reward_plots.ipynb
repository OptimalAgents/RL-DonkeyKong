```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import MaxNLocator
```

```python
data = np.load('eval_rewards/AH.npy')
y = data.flatten()

y_series = pd.Series(y)
window_size = 10
moving_avg = y_series.rolling(window=window_size).mean()

x = range(len(y))
x_moving_avg = [i for i in range(len(moving_avg))]

plt.plot(x, y, label='Original Data', alpha=0.75)

plt.plot(x, moving_avg, label='Moving Average (100k steps)', color='red')

plt.xlabel('Steps [x10000]')
plt.ylabel('Reward')
plt.title('Reward value over time')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0, 7))
plt.legend()
plt.xlim(xmin=0, xmax=len(y)-1)
plt.show()
```
